```python
df_train = df.sample(frac=0.85, random_state=42)
df_test = df.drop(df_train.index)
len(df_train), len(df_test)
```

```python
MODEL_NAME = 'aubmindlab/bert-base-arabertv01'
```

```python
t = text.Transformer(MODEL_NAME)
```

```python
trn = t.preprocess_train(df_train.text.values, df_train.label.values)
val = t.preprocess_test(df_test.text.values, df_test.label.values)
```

```python
model = t.get_classifier()
```

```python
learner = ktrain.get_learner(model, train_data=trn, val_data=val, batch_size=64)
learner.fit_onecycle(5e-5, 5)
```

```python
# Save the trained model and checkpoints
model_path = '/kaggle/working/my_model'
learner.save_model(model_path)
```

```python
learner.validate(class_names=t.get_classes())
```
